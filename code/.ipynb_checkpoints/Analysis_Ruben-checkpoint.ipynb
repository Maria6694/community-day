```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.api as sm
```

```python
cholera = pd.read_csv('../data/cholera_clean.csv', index_col=0)
cholera.head()
```

```python
cholera[['gastric_acid','vomit_col','fec_cons']].corr()
```

# Let's answer to the question:

# Is there any correlation between the gastric acid index and the fecen consistency and vomit color?


```python
X1 = cholera.fec_cons
X2 = cholera.vomit_col
Y = cholera.gastric_acid
syntoms = ['fec_cons', 'vomit_col']
all_syntoms = cholera.columns
##Generatio of intercept for all independent parameters
X = sm.add_constant(cholera[syntoms])

##Creation an object type OLS ordinary Least Squares
model = sm.OLS(Y,X)

##make the calculus for fitting the optimal linear regression
results = model.fit()
print(results.summary())
```

# Let's check  with all of parameters, how many can be reject the null hypohtesis, and consequently which coefficientes are non-zero values.

```python
Y = cholera.gastric_acid
syntoms = list(cholera.columns)
syntoms.remove('cholera')
syntoms.remove('gastric_acid')
syntoms.remove('sex')

##Generatio of intercept for all independent parameters
X = sm.add_constant(cholera[syntoms])

##Creation an object type OLS ordinary Least Squares
model = sm.OLS(Y,X)

##make the calculus for fitting the optimal linear regression
results = model.fit()
print(results.summary())
```

## if P>|t| > 0.05, the null hypothesis can not be rejected. Therefore the coefficient of this parameter is zero.

### Remove 'blood_pres', 'cholesterol', 'seizures', 'vomit_col'

```python
X1 = cholera.fec_cons
X2 = cholera.vomit_col
Y = cholera.gastric_acid
syntoms = list(cholera.columns)
syntoms.remove('cholera')
syntoms.remove('gastric_acid')
syntoms.remove('sex')
syntoms.remove('blood_pres')
syntoms.remove('cholesterol')
syntoms.remove('seizures')
syntoms.remove('vomit_col')



##Generatio of intercept for all independent parameters
X = sm.add_constant(cholera[syntoms])

##Creation an object type OLS ordinary Least Squares
model = sm.OLS(Y,X)

##make the calculus for fitting the optimal linear regression
results = model.fit()
print(results.summary())
```

# This is the best fit with all parameters with null hyphotesis can be rejected.
